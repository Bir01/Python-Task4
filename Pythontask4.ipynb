```python
#####Task4 Sözlügü Dosyalama
"""
Aşağıdaki program
'opposite.txt' isimli bir dosyada 
bir kelimenin zit anlam bilgilerini tutuyor.

"""
#Menü fonksiyonu
def menuSecim():
    print('[1]. Listele')
    print('[2]. Ekle')
    print('[3]. Sil')
    print('[4]. Ara')
    print('[5]. Güncelle')
    print('[6]. Çıkış')
    menuSec = 0
    while menuSec != 6:
        menuSec = int(input("Seçiminiz: "))
        if menuSec == 1:
            listele()
        elif menuSec == 2:
            ekle() 
        elif menuSec == 3:
            ad = input("Silinecek kelime.:")
            sil(ad)
        elif menuSec == 4:
            ad = input("Aranacak kelime.:")
            ara(ad)
        elif menuSec == 5:
            ad = input("Güncellenecek kelime.:")
            guncelle(ad)
        elif menuSec == 6:
            cikis() 
        else:
            print("1-6 arasında seçim yapınız!!")

def dosyaYaz(Kelime ='', ZitAnlam=''):
    f = open("opposite.txt",'a') #Dosyayı ekleme modunda aç
    f.write(Kelime+':\t'+ ZitAnlam +'\n')
    f.close()
#Kelime ekleme fonksiyonu
def ekle():
    Kelime=input('Kelime:')
    ZitAnlam=input('ZitAnlam:')
    dosyaYaz(Kelime, ZitAnlam) 
#Dosyadan kelime okuma fonksiyonu
def dosyaOku():
    f = open("opposite.txt",'r')  #Dosyayı okuma modunda aç
    Kelimeler = f.readlines()#Bütün satırları oku
    f.close() #dosyayı kapat
    return Kelimeler #Okunan tüm kayitları döndür
#Dosyadaki tüm kelimeleri listeleyen fonksiyon
def listele():
    klm=dosyaOku()
    for i in klm:
        print(i)
#Kelimeye göre arama yapan fonksiyon 
def ara(Kelime):
    global flag #kelime varsa flag=1, değilse flag =0 olacak
    try:
        f=open("opposite.txt",'r')#okuma amaçlı aç
        f.seek(0) #dosya başına konumlan
        Kelimeler=f.readline() #satır satır oku ve değişkene aktar 
        while Kelimeler !='': #kelimelerin sonuna gelinene kadar
            if Kelimeler.find(Kelime)!=-1: #eğer kelime kelimelerde varsa
                print (Kelimeler) #o adı yaz
                flag=1
                #  Arananın pozisyonu :
                #  Dosya başlangıç adresi (f.tell) - (kelime uzunluğu+1)
                return int(f.tell())-int(len(Kelimeler)+1) 
            else:
                Kelimeler=f.readline()
                flag=0
    except Exception: 
        print("Dosya Okuma hatası")
    finally:
        f.close()
 
    if flag==0:
        print ('Aranan bulunamadı') #hiçbir kelime yok ise
        return -1 # -1 değerini döndür

#Kelimeye göre silme yapan fonksiyon 
def sil(Kelime):
    p = ara(Kelime) #Silinecek kelimeyi ara
    if p!=-1: #kayıt mevcutsa
        st=dosyaOku()
        f=open("opposite.txt",'w')#yazma amaçlı aç
        f.writelines(st) 
 
        f.seek(p) #silinecek kelimeye konumlan
        f.write('*')
    else:
        print ('Silinecek kelime yok')#kelime dosyada yoksa

#Kelimeye göre düzeltme yapan fonksiyon
def guncelle(Kelime):
    p = ara(Kelime)
    if p!=-1: 
        st=dosyaOku()
        f=open("opposite.txt",'w')
        f.writelines(st)
        f.seek(p)
        #düzeltilecek kelimeler konumlan ve o kelimeler üzerine yeni bilgileri yaz
        try:
            Kelime = input('Kelime:')
            ZitAnlam = input('ZitAnlam:')
            
            f.write(Kelime+':\t'+ZitAnlam+'\n')
        except Exception: 
            print('Dosyaya yazma hatası...')
 
        else:
            print ('Böyle bir kelime yok..')#kelime dosyada yoksa
    f.close()
#Çıkış fonksiyonu
def cikis():
    print("Program sonlandırıldı")
    raise SystemExit() #programdan çık
#Ana program 
if __name__ == "__main__":
    dosya = "opposite.txt"   #dosya adı değişkeni
    menuSecim()             #menü fonksiyonunu çağır
```
